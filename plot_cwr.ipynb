```python
rfrom scipy.io import loadmat
import matplotlib.pyplot as plt
```

```python
baseline = loadmat("97.mat")
faulty = loadmat("105.mat") # 12k Drive End Bearing Fault Data
```

```python
de_b = baseline["X097_DE_time"]
de_f = faulty["X105_DE_time"]
print("Baseline length of data: {}".format(len(de_b)))
print("Fault length of data: {}".format(len(de_f)))
```

```python
minr=0
maxr=1000
plt.figure(figsize=(15,6))
plt.plot(de_f[minr:maxr], label = "Faulty")
plt.plot(de_b[minr:maxr], label = "Baseline")
plt.legend(loc='best')
plt.show
```
